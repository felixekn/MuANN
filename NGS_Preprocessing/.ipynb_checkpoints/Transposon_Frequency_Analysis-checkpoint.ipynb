```python
#modified to center everything on residue insted of left adjusted 
import csv, numpy

#Generate 75bp sequences from circular AK gene 
gene_fwd = "ATGAATCTGGTGCTGATGGGCCTGCCGGGTGCGGGCAAAGGCACCCAGGGCGAACGTATTGTGGAAGATTATGGAATTCCGCATATTAGCACCGGCGATATGTTTCGTGCGGCGATGAAAGAAGAAACCCCGCTGGGCCTGGAAGCGAAAAGCTATATTGATAAAGGCGAACTGGTGCCGGATGAGGTGACCATTGGCATTGTTAAAGAACGTCTGGGCAAAGATGATTGCGAACGCGGCTTTCTGCTGGATGGCTTTCCGCGGACCGTTGCGCAGGCGGAAGCACTGGAAGAAATTCTGGAAGAATACGGCAAACCGATTGATTACGTAATTAACATTGAAGTGGATAAAGATGTTCTGATGGAGCGCTTAACCGGCCGTCGTATTTGCAGCGTGTGCGGCACCACCTATCATCTGGTGTTTAACCCGCCGAAAACCCCGGGCATTTGCGATAAAGATGGCGGCGAACTGTATCAGCGCGCGGATGATAACGAAGAAACCGTTAGCAAACGTCTGGAAGTGAACATGAAACAGACCCAGCCGCTGCTGGATTTTTATAGCGAAAAAGGGTACCTGGCGAACGTTAACGGCCAGCAGGATATTCAGGATGTTTATGCGGATGTTAAAGATCTGCTGGGCGGCCTGAAAAAAGCGGCCGCA"
gene_rc  = "TGCGGCCGCTTTTTTCAGGCCGCCCAGCAGATCTTTAACATCCGCATAAACATCCTGAATATCCTGCTGGCCGTTAACGTTCGCCAGGTACCCTTTTTCGCTATAAAAATCCAGCAGCGGCTGGGTCTGTTTCATGTTCACTTCCAGACGTTTGCTAACGGTTTCTTCGTTATCATCCGCGCGCTGATACAGTTCGCCGCCATCTTTATCGCAAATGCCCGGGGTTTTCGGCGGGTTAAACACCAGATGATAGGTGGTGCCGCACACGCTGCAAATACGACGGCCGGTTAAGCGCTCCATCAGAACATCTTTATCCACTTCAATGTTAATTACGTAATCAATCGGTTTGCCGTATTCTTCCAGAATTTCTTCCAGTGCTTCCGCCTGCGCAACGGTCCGCGGAAAGCCATCCAGCAGAAAGCCGCGTTCGCAATCATCTTTGCCCAGACGTTCTTTAACAATGCCAATGGTCACCTCATCCGGCACCAGTTCGCCTTTATCAATATAGCTTTTCGCTTCCAGGCCCAGCGGGGTTTCTTCTTTCATCGCCGCACGAAACATATCGCCGGTGCTAATATGCGGAATTCCATAATCTTCCACAATACGTTCGCCCTGGGTGCCTTTGCCCGCACCCGGCAGGCCCATCAGCACCAGATTCAT"

gene_fwd = list(gene_fwd)
gene_rc = list(gene_rc)

fragments_F = []
fivemer_F = []
elevenmer_F = []

#fill forward frame
for seq in [gene_fwd]:
    for i in range(0,len(seq)):
        if i < 34:
            index_adjustment = len(seq)-(34-i)
            sample = seq[index_adjustment:len(seq)] + seq[0:i+35]
            fragments_F.append(''.join(sample))
        if 34 <= i < (len(seq)-34):
            sample = seq[i-34:i+35]
            fragments_F.append(''.join(sample))
        if i >= (len(seq)-34):
            index_adjustment = 34-(len(seq)-i)+1
            sample = seq[i-34:len(seq)] + seq[0:index_adjustment]
            fragments_F.append(''.join(sample))
       
        #Pull out central 5mer and 11mers to match to corresponding transposon insertion frequencies 
        if i < 2:
            index_adjustment = len(seq) - (2-i)
            sample = seq[index_adjustment:len(seq)] + seq[0:i+2+1]
            fivemer_F.append(''.join(sample)) 
        if 2 <= i < (len(seq)-2):
            sample = seq[i-2:i+3]
            fivemer_F.append(''.join(sample)) 
        if i >= (len(seq)-2):
            index_adjustment = (2-(len(seq)-i))
            sample = seq[i-2:len(seq)] + seq[0:index_adjustment+1]
            fivemer_F.append(''.join(sample)) 
        if i < 5:
            index_adjustment = len(seq) - (5-i)
            sample = seq[index_adjustment:len(seq)] + seq[0:i+5+1]
            elevenmer_F.append(''.join(sample)) 
        if 5 <= i < (len(seq)-5):
            sample = seq[i-5:i+5+1]
            elevenmer_F.append(''.join(sample)) 
        if i >= (len(seq)-5):
            index_adjustment = (5-(len(seq)-i))
            sample = seq[i-5:len(seq)] + seq[0:index_adjustment+1]
            elevenmer_F.append(''.join(sample)) 

            
print(fragments_F[655:len(seq)])
print(elevenmer_F[655:len(seq)])
print(fivemer_F[655:len(seq)])

#fill reverse frame
fragments_NF = []
fivemer_NF = []
elevenmer_NF = []

#fill forward frame
for seq in [gene_rc]:
    for i in range(0,len(seq)):
        
        #69mer centered on residue 
        if i < 34:
            index_adjustment = len(seq)-(34-i)
            sample = seq[index_adjustment:len(seq)] + seq[0:i+35]
            fragments_NF.append(''.join(sample))
        if 34 <= i < (len(seq)-34):
            sample = seq[i-34:i+35]
            fragments_NF.append(''.join(sample))
        if i >= (len(seq)-34):
            index_adjustment = 34-(len(seq)-i)+1
            sample = seq[i-34:len(seq)] + seq[0:index_adjustment]
            fragments_NF.append(''.join(sample))
       
        #5mer cenetered on residue 
        if i < 2:
            index_adjustment = len(seq) - (2-i)
            sample = seq[index_adjustment:len(seq)] + seq[0:i+2+1]
            fivemer_NF.append(''.join(sample)) 
        if 2 <= i < (len(seq)-2):
            sample = seq[i-2:i+3]
            fivemer_NF.append(''.join(sample)) 
        if i >= (len(seq)-2):
            index_adjustment = (2-(len(seq)-i))
            sample = seq[i-2:len(seq)] + seq[0:index_adjustment+1]
            fivemer_NF.append(''.join(sample)) 
       
        #11mer centered on residue 
        if i < 5:
            index_adjustment = len(seq) - (5-i)
            sample = seq[index_adjustment:len(seq)] + seq[0:i+5+1]
            elevenmer_NF.append(''.join(sample)) 
        if 5 <= i < (len(seq)-5):
            sample = seq[i-5:i+5+1]
            elevenmer_NF.append(''.join(sample)) 
        if i >= (len(seq)-5):
            index_adjustment = (5-(len(seq)-i))
            sample = seq[i-5:len(seq)] + seq[0:index_adjustment+1]
            elevenmer_NF.append(''.join(sample)) 

fragments_NF = fragments_NF[::-1]
elevenmer_NF = elevenmer_NF[::-1]
fivemer_NF = fivemer_NF[::-1]
print(fragments_NF[35:40])
print(elevenmer_NF[35:40])
print(fivemer_NF[35:40])            
            
fragments = fragments_F + fragments_NF #reverse the reverse compliment 
fivemer = fivemer_F + fivemer_NF
elevenmer = elevenmer_F + elevenmer_NF

print('Number of Fragments (Fwd & Rev):')
print(len(fragments))
print('Number of Unique 5mers:')
print(len(numpy.unique(fivemer)))
print('Number of Unique 11mers:')
print(len(numpy.unique(elevenmer)))

rows = zip(range(1, len(fragments)+1), fragments, fivemer, elevenmer)
with open('sixtyninemers_BsAK_2.csv', 'wb') as f:
    writer = csv.writer(f)
    writer.writerow(['index', 'oligo', '5mer', '11mer'])
    for row in rows:
        writer.writerow(row)
        
print('Written to CSV')
        
```

```python
#Plot counts of unique fivemers
from plotly import __version__
import plotly.graph_objs as go
import numpy, scipy
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)

unique = numpy.unique(fivemer)
mer = []
count = []
for i in range(0, len(unique)):
    counter = 0
    for k in range(0, len(fivemer)):
        if fivemer[k] == unique[i]:
            counter = counter + 1
    count.append(counter)
    mer.append(unique[i])


data = [go.Bar(
            x=range(0, len(unique)),
            y=sorted(count)
    )]

layout = go.Layout(
    xaxis=dict(
        title='5mer',
        range=[0,len(unique)],
        showgrid=False,
        ),
    yaxis=dict(
        title = '5mer Counts in Gene',
        showgrid=False,
        ),
    bargap = 0
    )

fig = go.Figure(data = data, layout= layout)
iplot(fig)
```

```python
#Plot counts of unique fivemers
from plotly import __version__
import plotly.graph_objs as go
import numpy, scipy
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)

data = numpy.genfromtxt('testanalysis_tfa_BsAK_update.csv', delimiter = ',', names=True, dtype=None)

#NF = Rev throughout, F = Fwd throughout no longer matches functional/nonfunctional
bp_site_NF = []
elevenmer_NF_two = []
fivemer_NF_two = []
bp_site_F = []
elevenmer_F_two = []
fivemer_F_two = []
count = 0
count1 = 0
counter = 0
counter1 = 0
shift = 0

for i in range(0, len(data)-1):
    if data[i]['Tag'] == 'AR':
        shift = 8
    if data[i]['Tag'] == 'BF':
        shift = 8
    if data[i]['Tag'] == 'AF':
        shift = 2
    if data[i]['Tag'] == 'BR':
        shift = 2
    if data[i]['gene_strand'] == 'Rev':
        site = (data[i]['residue'] * 3) - (3 - data[i]['frame']) - shift - 1 #calculate nucleotide postion from residue + frame and shift by 2 for the center of the 5mer
        if data[i]['frame'] > 0: #eliminates reads not corresponding to circular TnAK
            if site < 0:
                print('kitty')
                site_update = (len(gene_fwd)-1) + site
                bp_site_NF.append(site_update) 
                elevenmer_NF_two.append(elevenmer_NF[site_update])
                fivemer_NF_two.append(fivemer_NF[site_update])
                if data[i]['5mer'] != fivemer_NF[site_update]:
                    count1 = count1 + 1
                    if count1 <= 10:
                        print('These do not match Rev')
                        print(site_update + 1) #adjust to match nucleotide labeleing indexed at 1
                        print(data[i]['Tag'])
                        print(data[i]['direction'])
                        print(data[i]['residue'])
                        print(data[i]['frame'])
                        print(data[i]['11mer'])
                        print(elevenmer_NF[site_update])
                        print(data[i]['5mer'])
                        print(fivemer_NF[site_update])
            if site >= 0:
                bp_site_NF.append(site) 
                elevenmer_NF_two.append(elevenmer_NF[site])
                fivemer_NF_two.append(fivemer_NF[site])
                if data[i]['5mer'] != fivemer_NF[site]:
                    count = count + 1
                    if count <= 10:
                        print('These do not match Rev')
                        print(site + 1) #adjust to match nucleotide labeleing indexed at 1
                        print(data[i]['Tag'])
                        print(data[i]['direction'])
                        print(data[i]['residue'])
                        print(data[i]['frame'])
                        print(data[i]['11mer'])
                        print(elevenmer_NF[site])
                        print(data[i]['5mer'])
                        print(fivemer_NF[site])

    if data[i]['gene_strand'] == 'Fwd':
        site = (data[i]['residue'] * 3) - (3 - data[i]['frame']) + shift - 1 #calculate nucleotide postion from residue + frame and shift by 2 for the center of the 5mer
        if data[i]['frame'] > 0: #eliminates reads not corresponding to circular TnAK
            if site <= (len(gene_fwd)-1):
                bp_site_F.append(site) 
                elevenmer_F_two.append(elevenmer_F[site])
                fivemer_F_two.append(fivemer_F[site])
                if data[i]['5mer'] != fivemer_F[site]:
                    counter = counter + 1
                    if counter <= 10:
                        print('These do not match Fwd')
                        print(site+1) #adjust to match nucleotide labeleing indexed at 1
                        print(data[i]['Tag'])
                        print(data[i]['direction'])
                        print(data[i]['residue'])
                        print(data[i]['frame'])
                        print(data[i]['11mer'])
                        print(elevenmer_F[site])
                        print(data[i]['5mer'])
                        print(fivemer_F[site])
            if site > (len(gene_fwd)-1):
                print('frog')
                site_update = (len(gene_fwd)-1)-site
                bp_site_F.append(site_update) 
                elevenmer_F_two.append(elevenmer_F[site_update])
                fivemer_F_two.append(fivemer_F[site_update])
                if data[i]['5mer'] != fivemer_F[site_update]:
                    counter1 = counter1 + 1
                    if counter1 <= 10:
                        print('These do not match Fwd')
                        print(site_update+1) #adjust to match nucleotide labeleing indexed at 1
                        print(data[i]['Tag'])
                        print(data[i]['direction'])
                        print(data[i]['residue'])
                        print(data[i]['frame'])
                        print(data[i]['11mer'])
                        print(elevenmer_F[site_update])
                        print(data[i]['5mer'])
                        print(fivemer_F[site_update])
                


print(len(bp_site_NF))
print(len(numpy.unique(bp_site_NF)))
print(bp_site_NF[0:5])
print(elevenmer_NF_two[0:5])
print(bp_site_F[0:5])
print(elevenmer_F_two[0:5])

print('cat')
print(len(numpy.unique(bp_site_NF)))
print(len(gene_fwd))
print(len(numpy.unique(bp_site_F)))

bp_hist_NF = numpy.histogram((numpy.asarray(bp_site_NF)), len(gene_fwd), range = (1,len(gene_fwd)))[0]
bp_hist_F = numpy.histogram(bp_site_F, len(gene_fwd), range = (1,len(gene_fwd)))[0]

bp_hist = numpy.concatenate((bp_hist_F, bp_hist_NF))

print(max(numpy.unique(bp_site_NF)))
print(min(numpy.unique(bp_site_NF)))

print(max(numpy.unique(bp_site_F)))
print(min(numpy.unique(bp_site_F)))


rows = zip(range(1, len(fragments)+1), fragments, fivemer, elevenmer, bp_hist)
with open('sixtyninemers_frequencies_BsAK_og2.csv', 'wb') as f:
    writer = csv.writer(f)
    writer.writerow(['index', 'oligo', '5mer', '11mer', 'TransposonCounts'])
    for row in rows:
        writer.writerow(row)
        
print('Written to CSV')
print(len(data))
print(count)
print(count1)
print(counter)
print(counter1)
```
